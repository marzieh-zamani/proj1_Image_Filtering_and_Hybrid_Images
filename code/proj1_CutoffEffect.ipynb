## Setup

```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from utils import vis_hybrid_image, load_image, save_image
from student_code import my_imfilter, create_hybrid_image
%load_ext autoreload
%autoreload 2

image1 = load_image('../data/dog.bmp')
image2 = load_image('../data/cat.bmp')

# display the dog and cat images
plt.figure(figsize=(3,3)); plt.imshow((image1*255).astype(np.uint8));
plt.figure(figsize=(3,3)); plt.imshow((image2*255).astype(np.uint8));
```

## Create filter

```python
cutoff_frequency = 7
filter = cv2.getGaussianKernel(ksize=cutoff_frequency*4+1,
                               sigma=cutoff_frequency)
filter = np.dot(filter, filter.T)

# let's take a look at the filter!
plt.figure(figsize=(4,4)); plt.imshow(filter);
```

## Apply filter to image
Both functions you need to implement in this project can be found in _student_\__code.py_. Start by implementing _my_\__imfilter_, which takes both a filter and an image, and returns the filtered image. This code block will use your _my_\__imfilter_ function to create and display a blurry version of the dog image.

```python
blurry_dog = my_imfilter(image1, filter)
plt.figure(); plt.imshow((blurry_dog*255).astype(np.uint8));
```

## Create hybrid image with default parameters (vis0)

```python
# Create hybrid image
low_frequencies, high_frequencies, hybrid_image = create_hybrid_image(image1, image2, filter)
vis0 = vis_hybrid_image(hybrid_image)

plt.figure(); plt.imshow((low_frequencies*255).astype(np.uint8));
plt.figure(); plt.imshow(((high_frequencies+0.5)*255).astype(np.uint8));

# Save results
save_image('../results/Cutoff_7_low_frequencies.jpg', low_frequencies)
save_image('../results/Cutoff_7_high_frequencies.jpg', high_frequencies+0.5)
save_image('../results/Cutoff_7_hybrid_image_scales.jpg', vis0)
```

## Create filter and hybrid image with filter size = 29 & cutoff_frequency =  4, 12

```python
# sigma = cutoff_frequency = 4

# Create filter
cutoff_frequency = 4
filter = cv2.getGaussianKernel(ksize=29,
                               sigma=cutoff_frequency)
filter = np.dot(filter, filter.T)

# Create hybrid image
low_frequencies, high_frequencies, hybrid_image = create_hybrid_image(image1, image2, filter)
vis_s4 = vis_hybrid_image(hybrid_image)

plt.figure(); plt.imshow((low_frequencies*255).astype(np.uint8));
plt.figure(); plt.imshow(((high_frequencies+0.5)*255).astype(np.uint8));

# Save results
save_image('../results/Cutoff_4_low_frequencies.jpg', low_frequencies)
save_image('../results/Cutoff_4_high_frequencies.jpg', high_frequencies+0.5)
save_image('../results/Cutoff_4_hybrid_image_scales.jpg', vis_s4)
```

```python
# sigma = cutoff_frequency = 12

# Create filter
cutoff_frequency = 12
filter = cv2.getGaussianKernel(ksize=29,
                               sigma=cutoff_frequency)
filter = np.dot(filter, filter.T)

# Create hybrid image
low_frequencies, high_frequencies, hybrid_image = create_hybrid_image(image1, image2, filter)
vis_s12 = vis_hybrid_image(hybrid_image)

plt.figure(); plt.imshow((low_frequencies*255).astype(np.uint8));
plt.figure(); plt.imshow(((high_frequencies+0.5)*255).astype(np.uint8));

# Save results
save_image('../results/Cutoff_12_low_frequencies.jpg', low_frequencies)
save_image('../results/Cutoff_12_high_frequencies.jpg', high_frequencies+0.5)
save_image('../results/Cutoff_12_hybrid_image_scales.jpg', vis_s12)
```
